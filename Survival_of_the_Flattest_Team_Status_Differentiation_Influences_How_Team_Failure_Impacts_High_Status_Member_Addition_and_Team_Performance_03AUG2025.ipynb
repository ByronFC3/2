‚ÄúSurvival of the Flattest: Team Status Differentiation Influences How Team Failure Impacts High-Status Member Addition and Team Performance‚Äù
By Yufei Zhong, Georgia Tech, May 2024

üìå Summary:
This dissertation investigates how teams respond to failure by altering their composition‚Äîspecifically, whether they add high-status members after poor performance‚Äîand whether this improves subsequent outcomes. Importantly, it highlights how team status differentiation (the extent to which power/status is concentrated in a few vs. shared more equally) moderates this relationship.

Using a massive longitudinal dataset of 25,000+ performance episodes across 5,283 product development teams in a large IT company, the study found:

Teams with low status differentiation are more likely to respond to failure by adding high-status members, which subsequently improves performance.

In contrast, highly hierarchical teams are less likely to add such members, likely due to status threat concerns among incumbents.

The addition of high-status members‚Äîrather than the loss or addition of lower-status ones‚Äîis the only membership change correlated with improved team performance post-failure.

üîë Key Takeaways:
1. Teams as Open Systems
Traditional research views teams as fixed groups. This work adopts a dynamic, open-system view, recognizing that membership changes can be strategic responses to failure.

2. High-Status Members as Valuable Additions
High-status individuals bring respected skills, knowledge, and networks that enhance team capability.

Their addition correlates with improved performance, but only when the team‚Äôs structure allows it.

3. Status Differentiation Moderates Recovery Behavior
In low-status-differentiated (flat) teams, failure increases the likelihood of adding high-status members.

In highly status-differentiated teams, self-interest and status-protection behaviors suppress such beneficial additions.

4. No Effect from Adding or Losing Low/Equal Status Members
The study tested numerous alternate explanations (e.g., loss of members, addition of low- or mid-status individuals) and found none improved performance.

5. Implications for Team Design
Flatter teams are more adaptable after failure.

Managers should encourage team-led, discretionary high-status additions, especially after setbacks.

Organizations should foster collective over individual incentives in hierarchical teams to reduce resistance to beneficial changes.

üíº Practical Implications:
Empower teams to select new members rather than imposing them.

Consider flattening team hierarchies to increase openness to performance-improving changes.

Encourage a shared identity and collective goals, especially in highly stratified teams.

üìö Theoretical Contributions:
Introduces a co-evolutionary view: Performance influences hierarchy, and vice versa.

Challenges the assumption that teams always act rationally to maximize collective goals.

Bridges team failure literature with status dynamics and social hierarchy research.
