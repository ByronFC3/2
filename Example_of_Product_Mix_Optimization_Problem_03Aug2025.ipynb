Example: Product Mix Optimization
Problem Statement:
A factory makes two products: Widgets and Gadgets. Each product requires time on two machines:

Machine A (hours)	Machine B (hours)	Profit per unit
Widget	2	1	$30
Gadget	1	2	$20

You have:

100 hours available on Machine A

80 hours available on Machine B

You want to determine how many Widgets (x) and Gadgets (y) to produce in order to maximize profit.

üìê Step 1: Define Decision Variables
Let x = number of Widgets to produce

Let y = number of Gadgets to produce

üìâ Step 2: Formulate the Constraints
Machine A time constraint:
2x + 1y ‚â§ 100

Machine B time constraint:
1x + 2y ‚â§ 80

Non-negativity constraints:
x ‚â• 0, y ‚â• 0

üí∞ Step 3: Define the Objective Function
Maximize total profit:
Maximize Z = 30x + 20y

üß† Final Linear Program:
Maximize
‚ÄÉ‚ÄÉZ = 30x + 20y

Subject to
‚ÄÉ‚ÄÉ2x + y ‚â§ 100
‚ÄÉ‚ÄÉ x + 2y ‚â§ 80
‚ÄÉ‚ÄÉx ‚â• 0, y ‚â• 0

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create x values
x = np.linspace(0, 60, 400)

# Constraint lines
y1 = 100 - 2*x         # From 2x + y = 100
y2 = (80 - x) / 2      # From x + 2y = 80

# Plot constraint lines
ax.plot(x, y1, label='2x + y ‚â§ 100')
ax.plot(x, y2, label='x + 2y ‚â§ 80')

# Fill feasible region
y1_clip = np.minimum(y1, y2)
y1_clip = np.maximum(y1_clip, 0)
ax.fill_between(x, 0, y1_clip, where=(y1_clip > 0), color='lightblue', alpha=0.5, label='Feasible Region')

# Plot corner points
points = {
    "A (0, 0)": (0, 0),
    "B (50, 0)": (50, 0),
    "C (40, 20)": (40, 20),
    "D (0, 40)": (0, 40)
}
for label, (px, py) in points.items():
    ax.plot(px, py, 'ro')
    ax.text(px + 1, py + 1, label, fontsize=9)

# Labels and title
ax.set_xlim(0, 60)
ax.set_ylim(0, 60)
ax.set_xlabel('Widgets (x)')
ax.set_ylabel('Gadgets (y)')
ax.set_title('Feasible Region and Constraint Lines')
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```
