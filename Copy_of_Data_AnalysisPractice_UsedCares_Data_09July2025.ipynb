Below is the price value per year the car is on the market

```python
us['Year'] = 2024 - us['Age']
us['price_per_year'] = us['Price'] / us['Year']
display(us.head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(us['price_per_year'], bins=30, kde=True)
plt.title('Distribution of Price Per Year')
plt.xlabel('Price Per Year')
plt.ylabel('Frequency')
plt.show()
```

Below:
The scatter plots show the relationship between price_per_year and KM, and price_per_year and HP.

From the plots, we can observe:

Price Per Year vs. KM: There seems to be a negative relationship between price_per_year and KM, meaning that as the kilometers driven increase, the price per year tends to decrease. This is expected as cars with higher mileage are generally worth less.
Price Per Year vs. HP: There doesn't seem to be a clear linear relationship between price_per_year and HP.

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='KM', y='price_per_year', data=us)
plt.title('Price Per Year vs. KM')
plt.xlabel('KM')
plt.ylabel('Price Per Year')

plt.subplot(1, 2, 2)
sns.scatterplot(x='HP', y='price_per_year', data=us)
plt.title('Price Per Year vs. HP')
plt.xlabel('HP')
plt.ylabel('Price Per Year')

plt.tight_layout()
plt.show()
```

```python
correlation_matrix = us.corr(numeric_only=True)
display(correlation_matrix)
```

```python
# Map out on a line diagram Price correleation to each varible

from matplotlib import pyplot as plt
correlation_matrix['Price'].plot(kind='line', figsize=(8, 4), title='Price')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# prompt: regression analysis

import pandas as pd
import statsmodels.api as sm

# Define the independent variables (features) and the dependent variable (target)
X = us[['Age', 'KM', 'HP']] # Using Age, KM, and HP as predictors
y = us['Price']

# Add a constant to the independent variables matrix (for the intercept)
X = sm.add_constant(X)

# Fit the OLS (Ordinary Least Squares) model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())

# You can also make predictions using the model
# For example, let's predict the price for a car with Age=5, KM=50000, HP=100
# new_data = pd.DataFrame({'const': 1, 'Age': [5], 'KM': [50000], 'HP': [100]})
# predicted_price = model.predict(new_data)
# print(f'\nPredicted Price: {predicted_price[0]}')
```
